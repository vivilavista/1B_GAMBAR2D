```python
#load the csv
data1 = pd.read_csv("nba.csv")

#show first 5 colomn
data1.head()
```

```python
data1.tail()
```

```python
sns.lineplot(data1["Salary"])
```

```python
sns.scatterplot(data1["Age"])
```

```python
sns.lineplot(x=data1["Age"], y=data1["Salary"])
```

```python
sns.lineplot(x=data1['Age'],y=data1['Weight'], hue =data1['Position'])
```

```python
sns.scatterplot(x=data1['Age'], y=data1['Weight'], hue = data1["Position"])
```

```python
sns.boxplot(x=data1['Age'],y=data1['Weight'])
```

```python
sns.boxplot(x=data1['Age'],y=data1['Weight'], hue=data1['Age'],palette=["blue","red"],legend=False)
```

```python
sns.pointplot(x = "Age", y = "Weight", data = data1)
```

```python
sns.barplot(x ="Age", y ="Weight", hue="Age",palette=["blue","red"],legend=False, data = data1)
```

```python
sns.barplot(x ="Age", y ="Weight", hue="Age",palette=["blue","red","green","pink"],legend=False, data = data1)
```
